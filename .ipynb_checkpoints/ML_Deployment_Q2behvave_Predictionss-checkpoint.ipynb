```python
# features to analyse
#mean
def feature_mean(feature_name,accel_gyro_data):
    mean = accel_gyro_data.mean()
    mean_transposed_df = mean.to_frame().T
    mean_transposed_df.columns = ["Accel_X_"+feature_name ,"Accel_Y_" +feature_name ,"Accel_Z_" +feature_name,"Gyro_X_" +feature_name, "Gyro_Y_" + feature_name ,"Gyro_Z_" +feature_name]
    return mean_transposed_df

#std
def feature_std(feature_name,accel_gyro_data):
    std = accel_gyro_data.std()
    std_transposed_df = std.to_frame().T
    std_transposed_df.columns = ["Accel_X_"+feature_name ,"Accel_Y_" +feature_name ,"Accel_Z_" +feature_name,"Gyro_X_" +feature_name, "Gyro_Y_" + feature_name ,"Gyro_Z_" +feature_name]
    return std_transposed_df

#kurtosis
def feature_kurtosis(feature_name,accel_gyro_data):
    kurtosis = accel_gyro_data.kurtosis()
    kurtosis_transposed_df = kurtosis.to_frame().T
    kurtosis_transposed_df.columns = ["Accel_X_"+feature_name ,"Accel_Y_" +feature_name ,"Accel_Z_" +feature_name,"Gyro_X_" +feature_name, "Gyro_Y_" + feature_name ,"Gyro_Z_"+feature_name]
    return kurtosis_transposed_df

#median
def feature_median(feature_name,accel_gyro_data):
    median = accel_gyro_data.median()
    median_transposed_df = median.to_frame().T
    median_transposed_df.columns = ["Accel_X_"+feature_name ,"Accel_Y_" +feature_name ,"Accel_Z_" +feature_name,"Gyro_X_" +feature_name, "Gyro_Y_" + feature_name ,"Gyro_Z_" + feature_name]
    return median_transposed_df

#skewness
def feature_skewness(feature_name,accel_gyro_data):
    skewness = accel_gyro_data.skew()
    skewness_transposed_df = skewness.to_frame().T
    skewness_transposed_df.columns = ["Accel_X_"+feature_name ,"Accel_Y_" +feature_name ,"Accel_Z_" +feature_name,"Gyro_X_" +feature_name, "Gyro_Y_" + feature_name ,"Gyro_Z_" + feature_name]
    return skewness_transposed_df
```

```python
# 'Accel_X_mean', 'Accel_Y_mean', 'Accel_Z_mean', 'Gyro_X_mean',
#        'Gyro_Y_mean', 'Gyro_Z_mean', 'Accel_X_std', 'Accel_Y_std',
#        'Accel_Z_std', 'Gyro_X_std', 'Gyro_Y_std', 'Gyro_Z_std',
#        'Accel_X_kurtosis', 'Accel_Y_kurtosis', 'Accel_Z_kurtosis',
#        'Gyro_X_kurtosis', 'Gyro_Y_kurtosis', 'Gyro_Z_kurtosis',
#        'Accel_X_skewness', 'Accel_Y_skewness', 'Accel_Z_skewness',
#        'Gyro_X_skewness', 'Gyro_Y_skewness', 'Gyro_Z_skewness', 'type'],
# create table 
def get_concated_features(accel_gyro_data):
    # Subset the DataFrame to include only the specified columns
    accel_gyro_data = accel_gyro_data[['Accel_X', 'Accel_Y', 'Accel_Z', 'Gyro_X', 'Gyro_Y', 'Gyro_Z']]
    df1 = feature_mean("mean",accel_gyro_data)
    df2 = feature_median("median",accel_gyro_data)
    df3 = feature_std("std",accel_gyro_data)
    df4 = feature_kurtosis("kurtosis",accel_gyro_data)
    df5 =feature_skewness("skewness",accel_gyro_data)
    df = [df1, df2, df3, df4, df5]
    result_df = pd.concat(df, axis=1)

    return result_df
```

```python

# Function to read accelerometer files and concatenate into one DataFrame
def read_accelerometer_files(directory):
    # List files in the directory
    files = os.listdir(directory)
    
    # Filter files with "acc" in their filename
    acc_files = [file for file in files if "acc" in file]
    
    # Initialize an empty list to store DataFrames
    dfs = []
    
    # Iterate over each accelerometer file
    for file in acc_files:
        # Read the file into a DataFrame
        df = pd.read_csv(os.path.join(directory, file))
        # Append the DataFrame to the list
        dfs.append(df)
    
    # Concatenate all DataFrames into one
    concatenated_df = pd.concat(dfs, ignore_index=True)
    
    return concatenated_df

```

```python

# Function to read accelerometer files and concatenate into one DataFrame
def read_gryoscope_files(directory):
    # List files in the directory
    files = os.listdir(directory)
    
    # Filter files with "acc" in their filename
    acc_files = [file for file in files if "gry" in file]
    
    # Initialize an empty list to store DataFrames
    dfs = []
    
    # Iterate over each accelerometer file
    for file in acc_files:
        # Read the file into a DataFrame
        df = pd.read_csv(os.path.join(directory, file))
        # Append the DataFrame to the list
        dfs.append(df)
    
    # Concatenate all DataFrames into one
    concatenated_df = pd.concat(dfs, ignore_index=True)
    
    return concatenated_df

```

```python
def merge_acc_gry(df1,df2):
    # Assuming df1 and df2 are your DataFrames with the given dimensions
    # Check which DataFrame has fewer rows
    if len(df1) < len(df2):
        smaller_df = df1
        larger_df = df2
    else:
        smaller_df = df2
        larger_df = df1

    # Truncate the larger DataFrame to match the number of rows in the smaller one
    truncated_larger_df = larger_df.iloc[:len(smaller_df)]

    # Concatenate the two DataFrames horizontally
    merged_df = pd.concat([smaller_df, truncated_larger_df], axis=1)
    # Assuming new_column_names is a list of new column names
    new_column_names = ['Accel_X', 'Accel_Y', 'Accel_Z', 'event.timestamp', 'Timestamp', 
                        'Gyro_X', 'Gyro_Y', 'Gyro_Z', 'event.timestamp2', 'Timestamp2']

    # Update the column names of the merged DataFrame/
    merged_df.columns = new_column_names
    # Define the new column order
    new_column_order = ['Accel_X', 'Accel_Y', 'Accel_Z', 
                        'Gyro_X', 'Gyro_Y', 'Gyro_Z', 
                        'event.timestamp', 'Timestamp', 
                        'event.timestamp2', 'Timestamp2']

    # Rearrange the columns of the DataFrame
    rearranged_df = merged_df.reindex(columns=new_column_order)

    return rearranged_df
    # Now, merged_df contains the two DataFrames merged horizontally with the same number of rows
```

```python
def make_predictions(svm_model,X_test):
 
    # Convert the DataFrame to a numpy array
    input_array = X_test.values
    # If necessary, reshape the input array to match the expected shape for prediction
    # For example, if your SVM model expects input shape (1, n_features), you can reshape it like this:
    input_array_reshaped = input_array.reshape(1, -1)

    # Make predictions using the loaded model
    predictions = svm_model.predict(input_array_reshaped)
    # Print the predictions
    print("Predictions:", predictions)
```

```python
def make_windows_of_each_seconds(accelerometer_data):
    # Load the SVM model from the .pkl file
    svm_model = joblib.load(model_file)
    # Convert the 'timestamp' column to datetime if it's not already in datetime format
    accelerometer_data['Timestamp'] = pd.to_datetime(accelerometer_data['Timestamp'])

    # Initialize an empty list to store windowed dataframes
    windowed_dataframes = []

    # Get the unique seconds values from the timestamp column
    unique_seconds = accelerometer_data['Timestamp'].dt.second.unique()
    print("unique_seconds: ",unique_seconds)
    # Iterate over each unique second value
    for second in unique_seconds:
        # Filter the data for the current second
        windowed_df = accelerometer_data[accelerometer_data['Timestamp'].dt.second == second]
        # Get features to run the ML model
        X_test_windowed_df_features = get_concated_features(windowed_df)
        X_test_windowed_df_features = X_test_windowed_df_features[['Accel_X_mean', 'Accel_Y_mean', 'Accel_Z_mean', 'Gyro_X_mean',
       'Gyro_Y_mean', 'Gyro_Z_mean', 'Accel_X_std', 'Accel_Y_std',
       'Accel_Z_std', 'Gyro_X_std', 'Gyro_Y_std', 'Gyro_Z_std',
       'Accel_X_kurtosis', 'Accel_Y_kurtosis', 'Accel_Z_kurtosis',
       'Gyro_X_kurtosis', 'Gyro_Y_kurtosis', 'Gyro_Z_kurtosis',
       'Accel_X_skewness', 'Accel_Y_skewness', 'Accel_Z_skewness',
       'Gyro_X_skewness', 'Gyro_Y_skewness', 'Gyro_Z_skewness']]
        make_predictions(svm_model,X_test_windowed_df_features)
        # Append the windowed dataframe to the list
        windowed_dataframes.append(windowed_df)
```
