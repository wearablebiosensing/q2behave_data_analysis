```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import scipy as sc

#from sklearn.preprocessing import MinMaxScaler

import os
```

```python
#scaler = MinMaxScaler()
from collections import defaultdict

# Define a dictionary to store the results
results = defaultdict(lambda: defaultdict(lambda: defaultdict(int)))

# read data in
# Specify the directory where your CSV files are located
directory_path = 'C:\\Users\\fried\\Documents\\URI\\ELE_591\\syncedfiles\\BlobsOFSegments'



for filename in os.listdir(directory_path):
    if filename.endswith('.csv') and 'Right' in filename:
        # Extract information from the filename
        parts = filename.split("_")
        p, a = parts[0], parts[2]
        
        # Extract hyperactive behavior and count information
        behavior = "_".join(parts[3:-1])  # Reconstruct the behavior and count part
        

        
        # Extract the count from the last part
        count = int(parts[-1].split(".")[0])
        #count = parts[-1]

        # Check if the current count is higher than the previous count for the same P and A combination
        if count > results[p][a][behavior]:
            results[p][a][behavior] = count


data = []

for p, a_dict in results.items():
    for a, behavior_dict in a_dict.items():
        row = {"Px": p, "Ax": a}
        row.update(behavior_dict)
        data.append(row)


df = pd.DataFrame(data)


df = df.fillna(0)


df.head()
#df.to_csv('C:\\Users\\fried\\Documents\\URI\\ELE_591\\syncedfiles\\frequency_of_behaviors.csv', index=False)
```

```python
sns.set(style="whitegrid")

# Get a list of unique activity names from the DataFrame
activity_names = df.columns[2:]  

# Create plots for each activity
for activity_name in activity_names:
    plt.figure(figsize=(10, 6))
    sns.barplot(x="Px", y=activity_name, data=df, hue="Ax")
    plt.xlabel("Participants (Px)")
    plt.ylabel("Count")
    plt.title(f"Hyperactive Behavior Counts for {activity_name}")
    plt.xticks(rotation=45)
    plt.legend(title="Ax")
    plt.show()
```

```python

A_values = df["Ax"].unique()

# Create a heatmap for each A
for A_value in A_values:
    plt.figure(figsize=(12, 8))
    subset_df = df[df["Ax"] == A_value]
    behavior_counts = subset_df.iloc[:, [0] + list(range(2, len(df.columns)-4))].set_index("Px")
    #behavior_counts = subset_df.iloc[:, :].set_index(["Px", "Ax"])  
    sns.heatmap(behavior_counts.T, annot=True, fmt=".1f", cmap="YlGnBu")
    plt.yticks(rotation=0)
    plt.title(f"Hyperactive Behavior Counts for {A_value}")
    
    
    plt.show()
```
